```python
import pandas as pd

#Nettoyer le fichier
df = pd.read_excel("ssr.xlsx", skiprows=3, index_col="Unnamed: 0")
df = df.rename(columns = {
    "Unnamed: 0" : "Catégorie"
})
df = df.replace("Premières diffusions", "Total")

#Ne prendre en compte que les premières diffusions
df = df[1:18]

#Ne prendre en compte que les années 1998 et 2019
df = df[[1998, 2019]]

#Calculer la différence en chiffres absolus entre 1998 et 2019 et le rajouter dans une colonne
df["Différence"] = df[2019] - df[1998]

#Calculer la différence en pourcentage entre 1998 et 2019 et la rajouter dans une colonne
df["Différence en %"] = (df["Différence"]/df[1998])*100

#Arrondir
df["Différence en %"] = round(df["Différence en %"], 2)
df[1998] = round(df[1998], 2)
df[2019] = round(df[2019], 2)
df["Différence"] = round(df["Différence"], 2)

#Afficher le résultat
df
```

```python
#Calculer la proportion de diffusion par catégorie sur le total pour chaque année
#Calculer pour chaque année le nombre d'heures de diffusion / le nombre total
proportion = 0
total_1998 = 32657.74

def calcul_pourcent(colonne):
    for line in df[colonne]:
        proportion = (line/total_1998)*100
        #print(round(proportion))
        return round(proportion)
    
df["% du programme en 1998"] = calcul_pourcent(1998)

df

#Stocker ce résultat dans deux nouvelles colonnes


#Afficher le résultat
```
