<br>Evolution de la politique de diffusion entre 1998 et 2019
<br><br><b>Comment y parvenir</b>
<br>Nettoyer le fichier
<br>Ne prendre en compte que les premières diffusions
<br>Ne prendre en compte que les années 1998 et 2019
<br>Calculer la différence en chiffres absolus entre 1998 et 2019 et la rajouter dans une colonne
<br>Calculer la différence en pourcentage entre 1998 et 2019 et la rajouter dans une colonne
<br>Arrondir

```python
import pandas as pd

#Nettoyer le fichier
df = pd.read_excel("ssr.xlsx", skiprows=3, index_col="Unnamed: 0")
df = df.rename(columns = {
    "Unnamed: 0" : "Catégorie"
})

#Ne prendre en compte que les premières diffusions
df = df[1:18]

#Ne prendre en compte que les années 1998 et 2019
df = df[ [1998, 2019]]

#Calculer la différence en chiffres absolus entre 1998 et 2019 et le rajouter dans une colonne
df["Différence"] = df[2019] - df[1998]

#Calculer la différence en pourcentage entre 1998 et 2019 et la rajouter dans une colonne
df["Différence en %"] = (df["Différence"]/df[1998])*100

#Arrondir
df["Différence en %"] = round(df["Différence en %"], 2)

#Afficher le résultat
df
```
